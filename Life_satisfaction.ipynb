```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
oecd = pd.read_csv('dataset/oecd.csv', index_col=None, na_values=['NA'])
```

```python
gdp = pd.read_csv('dataset/gdp.csv', thousands=',', delimiter='\t',
                             encoding='latin1', na_values="n/a")
```

```python
oecd = oced.pivot(index="Country", columns="Indicator", values="Value")
oecd.head()
```

```python
gdp.head()
```

```python
gdp.rename(columns={"2015": "GDP per capita"}, inplace=True)
gdp.set_index("Country", inplace=True)
full_stats = pd.merge(left=oecd, right=gdp, left_index=True, right_index=True)
full_stats.sort_values(by="GDP per capita", inplace=True)
full_stats
```
